# Using numpy to look for a correlation between time data and sea level rise
---

### Data Source
Global Average Absolute Sea Level Change, 1880-2014 from the US Environmental Protection Agency using data from CSIRO, 2015; NOAA, 2015.
https://datahub.io/core/sea-level-rise

The data describes annual sea levels from 1880 to 2013.  Measures are adjusted using two standards: Commonwealth Scientific and Industrial Research Organisation(CSIRO) and National Oceanic and Atmospheric Administration (NOAA)  

Raw Data file:  https://raw.githubusercontent.com/freeCodeCamp/boilerplate-sea-level-predictor/master/epa-sea-level.csv

For this exercise:
*  import the pandas library
*  import the numpy library
*  read the csv dataset containing data on sea-levels from the year 1880 to 2013 into a dataframe (df)
*  use df.head() and df.info() to inspect the data and the column data types



````python
import pandas as pd
import numpy as np
#from scipy.stats import linregress

def get_dataset(df):
  sealevel_df = df[(df['Year']>=1880) & (df['Year'] <=2013)]
  display(sealevel_df.head())
  df.info()
  #display(sealevel)
  return sealevel_df

url = 'https://raw.githubusercontent.com/freeCodeCamp/boilerplate-sea-level-predictor/master/epa-sea-level.csv'
sea_level_df = pd.read_csv(url)
get_dataset(sea_level_df)
````

3.  Calculate the Pearson product-moment correlation coefficient between year and the rise in sea level.  (*Expected output:  0.98 when rounded to 2 decimal places*)

````python
def cal_statistics(df,col):
  #Calculate some statistics on the level array, eg:
  nparray_col = np.array(df[col])
  # mean
  mean_col = np.mean(nparray_col)
  print('mean:',mean_col)
  # standard deviation
  stddev_col = np.std(nparray_col)
  print('std dev:',stddev_col)
  # total
  sum_col = np.sum(nparray_col)
  print('total :',sum_col)

  # the year with the biggest rise in level
  # the year with the lowest rise in level
  max_col = np.max(nparray_col)
  min_col = np.min(nparray_col)
  nparray_year = np.array(df['Year'])
  year_max = np.where(nparray_year == max_col) #df[df[col] == max_col]
  year_min = np.where(nparray_year == min_col) #df[df[col] == min_col]
  
  maxyear = np.where(nparray_col==max_col)
  #print(maxyear)
  #print(nparray_year[maxyear])
  minyear = np.where(nparray_col==min_col)
  #print(minyear)
  #print(nparray_year[minyear])
  print('the year with the biggest rise in level : ',nparray_year[maxyear][0])#['Year'])
  print('the year with the lowest rise in level : ',nparray_year[minyear][0])#['Year'])

def cal_coef(df,col1,col2):
  #Calculate the Pearson product-moment correlation coefficient between year and the rise in sea level.
  # (Expected output: 0.98 when rounded to 2 decimal places)
  nparray_col1 = np.array(df[col1])
  nparray_col2 = np.array(df[col2])
  coef = np.corrcoef(nparray_col1,nparray_col2).round(decimals=2)
  print(f'correlation coefficient for {col1} and {col2} :\n',coef)
  #return coef
  
cal_statistics(sea_level_df,'CSIRO Adjusted Sea Level')
cal_coef(sea_level_df, 'Year','CSIRO Adjusted Sea Level')
#print(coef)
````
